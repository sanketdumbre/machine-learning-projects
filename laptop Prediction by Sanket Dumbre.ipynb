```python
# Creating function to seprate intel core and other intel and amd processors from cpu name column.
def fetch_processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
```

```python
# Applying function and adding it into new column.
df['Cpu brand'] = df['Cpu Name'].apply(fetch_processor)
```

```python
def turn_memory_into_MB(value):
    if "GB" in value:
        return float(value[:value.find("GB")]) * 1000
    elif "TB" in value:
        return float(value[:value.find("TB")]) * 1000000
```

```python
df["Memory Amount"] = df["Memory Amount"].apply(turn_memory_into_MB)
```
